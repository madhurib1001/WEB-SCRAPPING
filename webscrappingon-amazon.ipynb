```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.amazon.in/s?k=realme&crid=105PZPLEO6VM1&sprefix=realm%2Caps%2C370&ref=nb_sb_noss_2")
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
names=soup.find("span",class_="a-size-medium a-color-base a-text-normal")
names
```

```python
names=soup.find_all("span",class_="a-size-medium a-color-base a-text-normal")
names
```

```python
#for finding mobile phones prices
prices=soup.find("span",class_="a-price-whole")
prices
```

```python
prices=soup.find_all("span",class_="a-price-whole")
prices
```

```python
descriptions=soup.find("div",class_="a-row a-size-base a-color-secondary s-align-children-center")
descriptions
```

```python
descriptions=soup.find_all("div",class_="a-row a-size-base a-color-secondary s-align-children-center")
descriptions
```

```python
ratings=soup.find("span",class_="a-size-base s-underline-text")
ratings
```

```python
ratings=soup.find_all("span",class_="a-size-base s-underline-text")
ratings
```

```python
namelist1=[]
for i in names:
    namelist1.append(i.text)
namelist1
```

```python
pricelist1=[]
for i in prices:
    pricelist1.append(i.text)
pricelist1
```

```python
descriptionlist1=[]
for i in descriptions:
    descriptionlist1.append(i.text)
descriptionlist1
```

```python
ratinglist1=[]
for i in ratings:
    ratinglist1.append(i.text)
ratinglist1
```

```python
import pandas as pd
web_series=pd.DataFrame({})
web_series["Names"]=namelist1
web_series["price"]=pricelist1
web_series["Description"]=descriptionlist1
web_series["Ratings"]=ratinglist1
```

```python
web_series
```

```python
web_series.to_csv("amazon.csv")
```
