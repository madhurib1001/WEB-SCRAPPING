```python
soup=BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
#for finding camera names
names=soup.find("div",class_="_4rR01T")
names
```

```python
#for finding all camera names
names=soup.find_all("div",class_="_4rR01T")
names
```

```python
#for finding camera prices
prices=soup.find("div",class_="_30jeq3 _1_WHN1")
prices
```

```python
#for finding all camera prices
prices=soup.find_all("div",class_="_30jeq3 _1_WHN1")
prices
```

```python
#for finding camera description
descriptions1=soup.find("ul",class_="_1xgFaf")
descriptions1
```

```python
#for finding camera all description
descriptions1=soup.find_all("ul",class_="_1xgFaf")
descriptions1
```

```python
#for finding camera ratings and reviews
ratings=soup.find("span",class_="_2_R_DZ")
ratings
```

```python
#for finding all camera ratings and reviews
ratings=soup.find_all("span",class_="_2_R_DZ")
ratings
```

```python
namelist1=[]
for i in names:
    namelist1.append(i.text)
namelist1
```

```python
pricelist1=[]
for i in prices:
    pricelist1.append(i.text)
pricelist1
```

```python
descriptionlist1=[]
for i in descriptions1:
    descriptionlist1.append(i.text)
descriptionlist1
```

```python
ratinglist1=[]
for i in ratings:
    ratinglist1.append(i.text)
ratinglist1
```

```python
import pandas as pd
web_series=pd.DataFrame({})
web_series["Description"]=descriptionlist1
web_series["Names"]=namelist1
web_series["price"]=pricelist1
web_series["Ratings"]=ratinglist1
```

```python
web_series
```

```python
web_series.to_csv("webscraponflipkart.csv")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=web_series['price'],y=web_series['Ratings'])
plt.title("Price vs rating")
```
